## __T-Test and P-Value Using Python__

## __T-Test__

## When is it used?
- When trying to compare the means of two groups to see if they differ
- When we want to compare the means of two samples
- When the sample size is very low, almost less than or equal to 30
- When the data follows a normal distribution

### There are three types of t-tests. They are:
1.   Independent sample t-test
2.   Paired sample t-test
3.   One sample t-test 



## 1. Independent sample t-test:
1. Checks the average of two independent, unrelated groups
2. Samples should be from two different populations

- Null hypothesis: $\mu_{a} = \mu_{b}$
- Alternate hypothesis: $\mu_{a} \ne \mu_{b}$

## 2. Paired sample t-test:
1. The average of two samples taken from the same population but at different points in time
- Null hypothesis: $\mu_{d}$ = 0
- Alternate hypothesis: $\mu_{d} \ne $ 0

## 3. One sample t-test:
1. The average of a single group is different from the known average
- Null hypothesis: $\mu_{a} =$ X
- Alternate hypothesis: $\mu_{a} \ne $ X

## Step 1: Create Sample Data

Let's import the necessary libraries and create sample data. We have two data sets: **a** and **b**.


```python
import random
random.seed(100)
a = [ random.gauss(50,20) for x in range(30)]
b = [ random.gauss(55,15) for x in range(30)]
```

## Step 2: Visualize the Data

Let's plot them on a seaborn plot.

```python
import seaborn as sns
sns.set_style('darkgrid')
sns.kdeplot(a,shade = True)
sns.kdeplot(b,shade = True)
```

**__Observation__**

This is the plot of **a** and **b**. The null hypothesis is: the mean of **a**
is equal to the mean of **b**.

## Step 3: Perform a Two-Sample Independent T-Test

We'll use the SciPy dot stats library, and we need to give equal variants that are equal to false.

__Note:__Â WeÂ knowÂ thatÂ theÂ variants
are not equal to **a** and **b** because the standard deviations
areÂ different.

```python
import scipy.stats as stats
```

```python
t_stat,p_value = stats.ttest_ind(a,b,equal_var=False)
```

```python
t_stat,p_value
```

__Observation__

Since the P value is greater than 0.05, we accept the null hypothesis. That is, the means for both distributions are almost the same.
Let's confirm this by using the code below: 

```python
import numpy as np
```

```python
np.mean(a), np.mean(b)
```

**__Observation__**

We can see that the mean values for both **a** and **b** are almost the same, and hence we accept the null hypothesis. So, this is an independent sample t-test.

##Â Step 4: Perform a Two-Sample Paired T-Test

- Here, the null hypothesis would be that the difference between the means is equal to zero.Â 
- An alternative hypothesis would be that the difference between the means isÂ notÂ equalÂ toÂ zero.Â 
Let's do it with a separate dataset.


```python
a = [ random.gauss(50,20) for x in range(30)]
b = [ random.gauss(60,25) for x in range(30)]
```

```python
t_stat,p_value = stats.ttest_rel(a,b)
```

```python
t_stat,p_value
```

__Observation__

Since the p-value is less than 0.05, we reject the null hypothesis because of the difference between the means isÂ notÂ equalÂ toÂ zero.Â 

Let's confirm this using the code below.

```python
np.mean(b) - np.mean(a)
```

__Observation__

The difference is definitely not zero, andÂ henceÂ weÂ rejectÂ theÂ nullÂ hypothesis.
