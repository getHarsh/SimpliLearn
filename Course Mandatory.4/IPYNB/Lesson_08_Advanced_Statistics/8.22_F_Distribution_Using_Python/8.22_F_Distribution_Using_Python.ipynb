## __F-Distribution Using Python__

The F-distribution is just the first sample's chi-square divided by its degree of freedom and the total divided by the second sample's chi-square divided by its degree of freedom.

  $ F \sim \frac {\chi_1^2/v1}{\chi_2^2/v2} $


Here, the degrees of freedom are v1 and v2.



Let's create the F-distribution with varying degrees of freedom and store it in data 1â€“4.

### Step 1: Import Libraries, and Create and Plot the F-Distributions


- Import the necessary libraries, and create F-distributions using scipy.stats
- Plot the F-distributions using Matplotlib


```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import f
x = np.linspace(0, 4.5, 1000)   
data_1 = f(1, 1, 0)
data_2 = f(5,8, 0)
data_3 = f(4, 4, 0)
data_4 = f(100, 200, 0)
# Plot F-distributions with various degrees of freedom
plt.figure(figsize=(12, 6),dpi = 150)
plt.plot(x,data_1.pdf(x), label = '1,1')
plt.plot(x, data_2.pdf(x), label = '5,8')
plt.plot(x, data_3.pdf(x), label = '4,4')
plt.plot(x, data_4.pdf(x), label = '100,200')
plt.legend()
```

__Observation__

- This is for degrees (1,1). 
- As the degrees of freedom increase, it approaches a normal distribution.
- ThisÂ isÂ an F-distribution.


### Step 2: Perform an F-Test

ThisÂ testÂ isÂ usedÂ toÂ testÂ ifÂ theÂ variantsÂ ofÂ theÂ twoÂ populationsÂ areÂ equal.

-Â TheÂ nullÂ hypothesisÂ isÂ theÂ variancesÂ areÂ equal:
$Â H_0Â :Â \sigma_1^2Â =Â \sigma_2^2Â $Â <br>

- The alternative hypothesis is the variancesÂ areÂ notÂ equal:
$Â H_1Â :Â \sigma_1^2Â \neÂ \sigma_2^2Â $


Below, we have created two sets of data with the same mean and standard deviation.

- Specify ddof as one to get the sample variance instead of the population variance
- Define the degree of freedom for the numerator and the denominator
-Â CalculateÂ theÂ p-value

```python
import random
random.seed(20)
```

```python
x = np.array([random.gauss(100,15) for x in range(20)])
y = np.array([random.gauss(100,15) for x in range(20)])
```

```python
f_test_stat = np.var(x,ddof=1) / np.var(y,ddof = 1)
```

```python
dfn = x.size - 1
dfd = y.size - 1
```

```python
p_value = 1 - f.cdf(f_test_stat,dfn,dfd)
```

```python
p_value
```

__Observation__

- We can see that the p-value is less than 0.05, which means that the null hypothesis is true.
- This means that there is actually aÂ relationshipÂ betweenÂ theseÂ variables.
-Â InÂ otherÂ words,Â theÂ variancesÂ areÂ equal.

Now, let's change this into a different set of distributions and perform the F-test.

```python
x = np.array([random.gauss(100,15) for x in range(20)])
y = np.array([random.gauss(200,15) for x in range(20)])
```

```python
f_test_stat = np.var(x,ddof=1) / np.var(y,ddof = 1)
dfn = x.size - 1
dfd = y.size - 1
p_value = 1 - f.cdf(f_test_stat,dfn,dfd)
```

```python
p_value
```

__Observation__

- We can see that the p-value here isÂ greaterÂ thanÂ 0.05.
- Hence, we need to reject theÂ nullÂ hypothesisÂ andÂ acceptÂ theÂ alternativeÂ hypothesis.
- This means that the variances areÂ notÂ equal.
