## __Chi-Squared Distribution Using Python__

- Chi-squared distribution is a continuous probability distribution.
- It is mainly used for hypothesis testing, the chi-square goodness field and the Chi-square test of independence.
- It is usually defined by the degree of independence.


Let's go ahead and plot some graphs to visualize the chi-squared distribution for varying degrees of freedom.

## Step 1: Import Libraries and Create Chi-Squared Distributions

- Import the necessary libraries and create chi-squared distributions of varying degrees of freedom
-  DF signifies the degree of freedom


```python
import numpy as np
```

```python
data1 = np.random.chisquare(df = 1,size = 1000)
data2 = np.random.chisquare(df = 2,size = 1000)
data3 = np.random.chisquare(df = 3,size = 1000)
# Print the first 10 elements of the data1 array
print(data1[:10])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## Step 2: Plot the Distributions

- A seaborn distplot shows a
histogram with a line on it.
- Since we don't want the histogram but only the line, we'll use hist equal to false.

```python
sns.distplot(data1, hist = False, label = 'dof 1')
sns.distplot(data2, hist = False, label = 'dof 2')
sns.distplot(data3, hist = False, label = 'dof 3')
plt.legend()
```

__Observation__

- This is the chi-squared distribution for various degrees of freedom.

- We can see that as the degree of freedom increases, the chi-squared distribution almost tends to become a normal distribution.

##  Step 3: Perform Chi-Squared Tests

- The chi-squared test is mainly used to check if there is a relationship between two samples of data.

### Step 3.1: Import the Libraries Needed and Create Sample Data

```python
from scipy.stats import chi2_contingency
```

```python
data = [[10,20,30],[6,9,17]]
```

###Step 3.2: Perform the Chi-Squared Test

- The null hypothesis is that there is no association between the variables.

- The alternative hypothesis is that there is an association or correlation between the two variables that we are testing for.

Let's calculate the chi-squared value and look at the p-value.



```python
stat, p_value, dof, chi_array = chi2_contingency(data)
```

```python
p_value
```

__Observation__

- The p-value is greater than 0.05. Hence, we need to reject the null hypothesis.

- This means that we need to accept the alternative hypothesis that exists between the groups that we're testing for.



### Step 3.3: Perform a Chi-Squared Test for a Smaller Value

Now, let's modify the previous data and plot the p-value.

```python
data = [[10,20,30],[9,1,8]]
```
