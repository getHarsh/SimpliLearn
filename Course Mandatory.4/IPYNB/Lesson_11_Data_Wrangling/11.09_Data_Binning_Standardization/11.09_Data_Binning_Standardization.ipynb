##__Data Binning: Standardization__

Here, we will standardize the data
using the Z-score, where the new value is calculated by the subtracting the average value from the current value and dividing this by the standard deviation.

Letâ€™s import the libraries and read the dataset.

### Step 1: Import the Required Libraries and Read the Dataset

- Import the pandas library
- Read the **HousePrices.csv** file into a DataFrame


```python
import pandas as pd
from scipy.stats import zscore
```

```python
df = pd.read_csv('HousePrices.csv')
```

```python
df.head()
```

__Observation__

- These are the first five rows of the dataset.

### Step 2: Calculate the Z-Score for the __Price__ Column

Before calculating the outliers in the **price** column, let's first create the Z-score for the prices.

- Import Z-score from scipy.stats
- Calculate the Z-scores for the __Price__ column, and add it to the DataFrame


```python
df['zscore-price'] = zscore(df['price'])
```

```python
df.head()
```

__Observation__

- Here, we can see that the Z-scores for the prices are created.

### Step 3: Detect and Mark the Outliers Using Z-score

Let's create a new column where we detect outliers.

- Import the NumPy library
- Set a threshold for detecting outliers as 2
- Mark outliers in the __Outliers__ column of the DataFrame using the calculated Z-scores


```python
import numpy as np
```

```python
threshold = 2
df['outliers'] = np.where((df['zscore-price'] - threshold > 0), True, np.where(df['zscore-price'] + threshold < 0,True,False))
```

```python
df.head()
```

__Observations__

- The price is greater than 2300000.

Now, let's use the describe function.

```python
df.describe()
```

__Observations__

- Here, the maximum price is 2600000, so it's almost
in that range.
- This is likely an outlier.


### Step 4: Drop Outliers from the DataFrame

Let's drop all the outliers in the dataset.

- Drop the rows containing outliers
- Display the resulting DataFrame


```python
df.drop((df[df['outliers']==True]).index,inplace = True)
```

```python
df
```

__Observation__

- Hence, we have a dataset that is free of all outliers.
