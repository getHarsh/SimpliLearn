## __Describing Data__

Here, let's learn about describing the data.


Let's first load the dataset.
- The df.head( ) method is used to print the first five rows.
- The df.tail( ) method is used to print the last five rows.
- The df.shape method is used to print the shape of the file.
- The df.column method is used to print the columns of the dataset.


### Step 1: Import the Required Libraries and Load the Dataset

- Import the pandas library
- Load the **HousePrices.csv** file


```python
import pandas as pd
```

```python
df = pd.read_csv('HousePrices.csv')
```

### Step 2: Utilize Different Methods to Print the Dataset

Here, let's use the df.head( ), df.tail( ), df.shape and df.column method on the dataset.


```python
df.head()
```

__Observation__

- Here, the first five rows of the dataset are printed.

```python
df.tail()
```

__Observation__

- Here, the last five rows of the dataset are printed.

```python
df.shape
```

__Observation__

- The dataset has 4600 rows and 18 columns.

```python
df.columns
```

__Observation__

- Here, all the columns in the dataset are printed.


## Step 3: Get a Statistical Overview of the Dataset
To get an overview
of the dataset, use the describe function:

```python
df.describe()
```

__Observations__

Here, we have received the outputs given below:

1. The overall count, which is the number of rows
2. The mean of the values
3. The standard deviation
4. The minimum and maximum values
5. The values at the 25th percentile,
50th percentile and 75th percentile




## Step 4: Add Custom Percentiles to the Statistical Summary

 - The custom percentiles can also be added in the described function.
 - The values range from 0 to 1. 

```python
df.describe(percentiles=[0.1,0.3,0.5,0.7,0.9])
```

__Observation__

- This gives the values
of the percentiles at 10%, 30%, 50%, 70% and 90%.

## Step 5: View the Data Types of the Columns

Now, let's print the data types by using the dtypes method.



```python
df.dtypes
```

__Observation__

- The data types of the individual rows and columns are printed.
- For example, the data type of the **date** column is **object** and the data type of the **price** column is **float**.
- It automatically type-casts or infers the data types.

This is how we describe data in pandas.
