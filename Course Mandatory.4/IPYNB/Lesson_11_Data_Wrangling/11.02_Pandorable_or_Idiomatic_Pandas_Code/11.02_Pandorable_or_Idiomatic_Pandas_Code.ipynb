## Pandorable or Idiomatic Pandas Code

Pandas have their own set of idioms. They are referred to as pandorable by the developers.

## Step 1: Import the Required Library

Let's talk about pandorable and idiomatic pandas code. We'll be looking at the indexing method, chaining memory optimizations, groupby and also plotting. 

- Import the pandas library:


```python
import pandas as pd
```

## Step 2: Load the Dataset

- Load the __HousePrices.csv__ dataset using pandas:


```python
df = pd.read_csv('HousePrices.csv')
```

```python
df.head() 
```

## Step 4: Check the DataFrame Size and Memory Usage

Let's look at memory optimization. First, Let's see how much memory the DataFrame is consuming.


Let's see the individual splits of the columns by using memory_usage().

```python
df.memory_usage()
```

**Observation**

We can see the memory occupied by each column.

We can reduce the size of memory usage by manipulating certain columns, such as changing the datatypes of the columns.

```python
df.head()
```

**Observation**

If we look at the head of the dataset, **bedrooms** cannot be in floating digits. 

Let's convert that to integer.

To convert the column values to integer, we can use astype('int32').

```python
df['bedrooms'] = df['bedrooms'].astype('int32')
```

```python
df.memory_usage()
```

__Observation__

Now, we can see the memory space occupied by the column **bedroom** is lower.

## Step 5: Groupby Operations

One of the very important functions in pandas that lets you group by a certain field is the groupby method. 

Let's go ahead and group by City. 

Inside the groupby, we need to specify the column and the aggregation method.
In this case, it will be mean.


- Perform groupby operations on the DataFrame 

- Calculate the mean based on the **city** column

```python
df.groupby(['city']).mean()
```

**Observation**

Here, we can see the mean values of each city.

Now, let's find the maximum value in this city data.

```python
df.groupby(['city']).max()
```

**Observation**

Here, we can see the max values of each city.

## Step 6: DateTime Operations


Let's set an index. Here, we need to convert the **date** column into datetime values and set that as an index.


- Convert the **date** column to datetime using pd.datetime()
- Reassign this to df['date'] 

```python
df['date'] = pd.to_datetime(df['date'])
```

- Set the index of the DataFrame to the 'date' column using set_index()

- Modify the same DataFrame by passing **inplace = True**

```python
df.set_index(['date'],inplace = True)
```

Let's see the first few rows of the DataFrame.

```python
df.head()
```

**Observation**

We can see the index is changed to datetime.

Before we plot date, time, values or time series, data can be resampled, which is a form of groupby using the resample() method.

Resample the DataFrame at different time intervals, and calculate the mean:



Since this is daily data, we can convert this into seven days.

```python
df.resample('7D').mean()
```

```python
df.resample('M').mean()
```

```python
df.resample('Y').mean()
```
