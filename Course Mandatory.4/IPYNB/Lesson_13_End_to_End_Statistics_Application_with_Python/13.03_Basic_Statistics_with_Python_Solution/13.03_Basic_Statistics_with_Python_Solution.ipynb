## Step 1: Import Required Libraries and Read the Dataset

Import pandas, matplotlib.pyplot, seaborn, and warnings libraries


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_style('darkgrid')

import warnings
warnings.filterwarnings('ignore')
```

Read the `pivot.csv` file

```python
df = pd.read_csv('pivot.csv')
```

Display the first 5 rows

Use shape method to identify the number of rows and column in the dataset

## Step 2: Preprocess the Dataset

As we need to do the weekly and monthly analysis of the data, let's convert the **Date** column to the **DateTime** format and set it as the index.


```python
df['Date'] = pd.to_datetime(df['Date'])
df.set_index(['Date'],inplace = True)
```

```python
df.head()
```

## Step 3: Check for Distribution and Skewness Using Plot

- Plot the distribution of each column using seaborn


```python
for col in df.columns:
    sns.distplot(df[col])
    plt.show()
```

## Step 4: Describe the Dataset and Analyze Drug Prices

- Describe the dataset


```python
df.describe()
```

**Inferences from the describe function:**
1. Most number of units sold were of Drug4.
2. Least number of units sold were of Drug6.
3. Most sales happened with Drug4 while Drug2 contributed the least.
4. Even though Drug6 sold a max of 3 units, it generated 30000 worth of revenue implying that the price of each drug is not the same.

Let's check the cost of each drug.

```python
import numpy as np
```

Now, check for drug1

```python
np.mean(df['Drug1_Sales']/df['Drug1'])
```

*Check* now for other drugs as well

```python
drugs = ['Drug'+str(x) for x in range(1,9)]
for drug in drugs:
    print(drug,end = ' - ')
    print(np.mean(df[drug+'_Sales']/df[drug]))
```

**Inferences of drug price:**
1. No price fluctuation between weeks
2. All drugs have different prices.
3. Confirms our speculation that Drug6 is more costly than the others.
4. Drug6 is the costliest while Drug5 is the cheapest.


## Step 5: Perform Weekly Analysis

- Resample the dataset for a week and call it wdf
- Plot the weekly sales overview for each drug


```python
wdf = df.resample('W').mean()
```

```python
wdf.head()
```

```python
wdf.tail()
```

```python
wdf.describe()
```

**Inferences from the describe function of weekly analysis**:
1. Drug4 tops the sales.
2. Drug6 sold the least.
3. Drug7 performed the worst in terms of sales even if it sold close to 6 units.

Let's visualize the Drug1 on weekly basis.

```python
plt.figure(figsize=(6,3), dpi = 100)
plt.tick_params(axis = 'x', rotation = 45)
plt.title('Drug1 Weekly Sales Overview')
plt.bar(x = wdf.index, height = wdf.Drug1_Sales)
```

**Inferences from Drug1 weekly sales:**
1. The best week was from 17/09 to 21/09.
2. Sales picked up gradually with hiccups, topped at 21/09, and then saw a decline.
3. First and last week of September made almost the same sales.

Let's visualize the Drug2 on weekly basis.
