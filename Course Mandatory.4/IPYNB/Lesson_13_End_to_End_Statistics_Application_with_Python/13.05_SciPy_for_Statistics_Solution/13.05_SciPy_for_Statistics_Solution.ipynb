## __SciPy for Statistics Solution__

## Step 1: Import Required Libraries and Configure Settings

- Import numpy, matplotlib.pyplot, seaborn, scipy.stats, and scipy.optimize libraries
- Configure warnings and matplotlib settings


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.stats import norm
from scipy.optimize import curve_fit

import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

## Step 2: Generate the Data

Calculate the steps that are required to calculate data in the **numpy.arrange()** function

- Display the length, mean, and standard deviation of the array


```python
steps = (200-90)/1100
print(steps)
```

**Observation:**
- The steps would be 0.1 since 1100 data points need to be generated.

Generate an array of data using the below steps

```python
arr = np.arange(90,200,steps)
print(arr)
```

## Step 3: Plot the Data

Let's plot a line graph of the array.


```python
plt.figure(figsize = (10,7), dpi = 100)
plt.title('Line Plot of the Array')
plt.plot(arr)
```

Calculate the length of the array

```python
len(arr)
```

**Length of the data is** 1100.

Calculate the mean of the array

```python
np.mean(arr)
```

**Mean of the data is** ~ 145.

Calculate the standard deviation of the data

```python
np.std(arr)
```

**Standard Deviation of the data is 31.75**.

**Note:** Based on the plot above, which is linear and doesn't provide any valuable information, let's create a random variance.

## Step 4: Generate Random Samples and Create a Scatter Plot

Generate random samples



```python
data = norm.rvs(arr)
```

Let's create axis on which we can plot the data using `np.linspace()`.

```python
axis = np.linspace(1,1100,1100)
```

Create a scatter plot of the random samples data

```python
# Scatter plot of axis and data
plt.figure(figsize = (10,7), dpi = 100)
plt.title("Scatter Plot")
sns.set_style('darkgrid')
sns.scatterplot(data)
```

## Step 5: Fit the Data Using a Linear Function

- Define a linear function
- Fit the data using the curve_fit function
- Display the optimized parameters and covariance matrix


```python
# Fitting function
# Straight line y = mx + c
def func(x,m,c):
    return m*x + c
```

```python
popt, pcov = curve_fit(func,axis,data, p0=[0,40])
```

```python
popt
```

```python
pcov
```

## Step 6: Plot the Fitted Line and the Original Data

- Calculate the predicted values using the optimized parameters
- Plot the fitted line and the original data


```python
y_pred = func(axis, popt[0], popt[1])
```

```python
plt.figure(figsize = (10,7) , dpi = 100 )
plt.title('Checking How Good Our Line Fits')
plt.scatter(data,axis)
plt.plot(y_pred,axis,color = 'r')
```

**We have created a good fit of the data**.

## Step 7: Plot the Distribution of the Predicted Values

- Plot the distribution of the predicted values using seaborn


```python
# So our y_pred follows a bell curve
sns.distplot(y_pred)
```
